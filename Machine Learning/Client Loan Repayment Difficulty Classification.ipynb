```python
pca = PCA(n_components=27)
pca.fit(df)  # Replace with your dataset

# Access the loadings matrix
loadings_matrix = pca.components_

# Identify the important attributes for each of the 28 principal components
for i in range(27):
    principal_component = loadings_matrix[i]
    # You can print or analyze the loadings for each attribute
    print(f"Principal Component {i+1} Loadings: {principal_component}")
```

```python
import numpy as np

# Assuming you have a 28x38 loadings matrix named 'loadings_matrix'
# Replace 'loadings_matrix' with your actual data
# Each row corresponds to a principal component

# Initialize a list to store the highest magnitude loading and its attribute number for each principal component
highest_magnitudes_and_attributes = []

# Iterate through each principal component (each row)
for i in range(loadings_matrix.shape[0]):
    principal_component_loadings = loadings_matrix[i]  # Get the loadings for a principal component
    highest_magnitude = np.max(np.abs(principal_component_loadings))  # Find the highest magnitude
    attribute_number = np.argmax(np.abs(principal_component_loadings))  # Find the attribute number with the highest magnitude
    highest_magnitudes_and_attributes.append((highest_magnitude, attribute_number))

# highest_magnitudes_and_attributes now contains tuples (magnitude, attribute_number) for each principal component
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve, auc
```

```python
def EvaluateModel(y_test,y_pred):
  report = classification_report(y_test, y_pred)
  print(report)
  cm = confusion_matrix(y_test, y_pred)
  plt.figure(figsize=(4, 4))
  sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, annot_kws={"size": 16})
  plt.xlabel('Predicted Labels', fontsize=16)
  plt.ylabel('True Labels', fontsize=16)
  plt.title('Confusion Matrix', fontsize=18)
  plt.show()
  fpr, tpr, thresholds = roc_curve(y_test, y_pred)
  roc_auc = auc(fpr, tpr)
  plt.figure(figsize=(5, 5))
  plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
  plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
  plt.xlim([0.0, 1.0])
  plt.ylim([0.0, 1.05])
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.title('Receiver Operating Characteristic (ROC) Curve')
  plt.legend(loc='lower right')
  plt.show()
```

**SelectKBest Method**

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2  # You can choose a different scoring function
```

```python
# Assuming 'X' is your feature matrix and 'y' is your target variable
# You can choose the number of top features ('k') you want to select
k = 37

# Create a SelectKBest object with the scoring function
# chi2 is commonly used for classification problems
selector = SelectKBest(score_func=chi2, k=k)
# Fit the selector to your data
X_new = selector.fit_transform(X, y)
# Use a list comprehension to get the names of the selected columns from the original DataFrame
selected_feature_indices = selector.get_support()

# Get the column names of the selected features
selected_feature_names = X.columns[selected_feature_indices]

# Now, selected_feature_names contains the names of the selected columns in X_new
print(selected_feature_names)
X_new = pd.DataFrame(X_new)
X_new.columns = X.columns[selected_feature_indices]
# Now, X_new contains only the top 'k' features selected by SelectKBest
```

**Gradient Boosting Classifier**

```python
#importing the necessary libraries
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
# transform data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
# train with Gradient Boosting algorithm
# compute the accuracy scores on train and validation sets when training with different learning rates

learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(X_train_scale, y_train)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(X_train_scale, y_train)))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2,0.5,0.75,1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Create GridSearchCV instance
grid_search = GridSearchCV(estimator=gb, param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1)

# Fit the GridSearchCV to the data
grid_search.fit(X_train_scale, y_train)
```

```python
# Get the best parameters and best estimator
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

# Print the best parameters
print("Best Parameters:", best_params)

# Use the best estimator to make predictions
y_pred = best_estimator.predict(X_test_scale)

# Print classification report on the test set
print("Classification Report on Test Set:")
print(classification_report(y_test, y_pred))
```
